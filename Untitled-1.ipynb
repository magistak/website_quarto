```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import asyncio
import random
```

```python

grid_size = 10
snake = [(5, 5)]
direction = "right"
food = (3, 3)
game_over = False
output = widgets.Output()

def render():
    with output:
        clear_output(wait=True)
        for y in range(grid_size):
            row = ""
            for x in range(grid_size):
                if (x, y) == food:
                    row += "🍎"
                elif (x, y) in snake:
                    row += "🟩"
                else:
                    row += "⬜"
            print(row)
        if game_over:
            print("\n💀 Game Over!")

def place_food():
    options = [(x, y) for x in range(grid_size) for y in range(grid_size) if (x, y) not in snake]
    return random.choice(options)

def move_snake():
    global snake, food, game_over
    head_x, head_y = snake[-1]
    if direction == "up": head_y -= 1
    if direction == "down": head_y += 1
    if direction == "left": head_x -= 1
    if direction == "right": head_x += 1
    new_head = (head_x, head_y)

    # Wall or self collision
    if (head_x < 0 or head_x >= grid_size or head_y < 0 or head_y >= grid_size or new_head in snake):
        game_over = True
        return

    snake.append(new_head)
    if new_head == food:
        food = place_food()
    else:
        snake.pop(0)

def on_direction_change(new_dir):
    global direction
    # Prevent reverse moves
    opposites = {"up": "down", "down": "up", "left": "right", "right": "left"}
    if new_dir != opposites.get(direction):
        direction = new_dir

# Buttons
up_btn = widgets.Button(description="↑")
down_btn = widgets.Button(description="↓")
left_btn = widgets.Button(description="←")
right_btn = widgets.Button(description="→")

up_btn.on_click(lambda _: on_direction_change("up"))
down_btn.on_click(lambda _: on_direction_change("down"))
left_btn.on_click(lambda _: on_direction_change("left"))
right_btn.on_click(lambda _: on_direction_change("right"))

controls = widgets.HBox([left_btn, up_btn, down_btn, right_btn])
```

```python
display(controls, output)

# Start game loop
async def game_loop():
    global game_over
    while not game_over:
        move_snake()
        render()
        await asyncio.sleep(0.5)

await game_loop()
```
